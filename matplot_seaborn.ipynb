```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2.3,6,1,0,6],marker='h',linestyle='dotted',ms=20,mec="red",mfc="black",linewidth=5)
plt.legend()
plt.show()
```

```python
#bar chart
products=["food",'clothings','electronic','home appliances']
quantity=[1200,45000,89000,30000]
plt.bar(products,quantity,color=['green','pink','black','red'])
plt.show()
```

```python
#pie chart
products=["food",'clothings','electronic','home appliances']
quantity=[1200,45000,89000,30000]
plt.pie(quantity,labels=products,colors=['green','pink','blue','red'],autopct='%1.1f%%')
plt.show()
```

```python
#histogram
marks=[35,36,50,75,92,98,93]
plt.hist(marks,color='red')
plt.show()
```

```python
import pandas as pd
t=pd.read_csv("/content/Titanic-Dataset.csv")
t
```

```python
plt.hist(t['Age'])
plt.show()
```

```python
plt.bar(t['Sex'],t['Survived'])
plt.xlabel("sex")
plt.ylabel("survived ")
plt.title("gender vs survived")
plt.show()
```

```python
#scatter plot
plt.scatter(t['Age'],t['Fare'])
plt.show()
```

```python
#box plot
plt.boxplot(t['Age'])
plt.show()
```

```python
print(t['Age'].isnull().sum())
print(t['Age'].describe())
```

```python
# Fill missing 'Age' values with the mean age
t['Age'].fillna(t['Age'].mean(), inplace=True)
print(t['Age'].isnull().sum()) # Verify no more missing values
```

Now that the missing 'Age' values are handled, let's plot the boxplot again.

```python
# Box plot of Age after handling missing values
t.boxplot(column='Age', by='Sex')
plt.title('Box Plot of Age by Sex (Missing Values Filled)')
plt.ylabel('Age')
plt.show()
```

```python
#area chart
plt.fill_between(t['Age'],t['Fare'])
plt.show()
```
